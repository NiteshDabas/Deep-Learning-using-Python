## Basic Artificial Neural Networks 

```python
# Import the libraries
import matplotlib
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
```

```python
# Generate sample dataset using Scikit Learn
np.random.seed(44)
features_set, labels = datasets.make_moons(500, noise=0.1)
labels=labels.reshape(labels.shape[0],1)
#dataset plot
plt.figure(figsize=(10,7))
plt.scatter(features_set[:,0], features_set[:,1], c=labels, cmap=plt.cm.winter)
```

```python
#Activation function and feedforward n/w

def sigmoid(x):
    return 1/(1+np.exp(-x))
def sigmoid_der(x):
    return (sigmoid(x))*(1-sigmoid(x))

#Weights declaration
wh = np.random.rand(len(features_set[0]),4)
wo = np.random.rand(4, 1)
#Learning rate declaration
lr = 0.5

#Feedforward and backpropagation
for epoch in range(75000):
    #Feedforward
    #Hidden layer
    zh=np.dot(features_set, wh)
    ah=sigmoid(zh)
    #Output layer
    zo=np.dot(ah, wo)
    ao=sigmoid(zo)
    
    #Loss function
    error=((1/2)*(np.power((ao-labels), 2)))
    print(error.sum())
    
    #Backpropagation
    #derivative cost-output layer
    dcost_dao=ao-labels
    #weights update-hidden layer
    dao_dzo=sigmoid_der(zo)
    #weights update-input layer
    dzo_dwo=ah
    #derivative cost function-input layer
    dcost_wo=np.dot(dzo_dwo.T, dcost_dao*dao_dzo)
    
    #Feedforward-again
    #derivative cost-final output
    dcost_dzo=dcost_dao*dao_dzo
    #output weights
    dzo_dah=wo
    #updated cost function
    dcost_dah=np.dot(dcost_dzo, dzo_dah.T)
    
    #Backpropagation-again
    #output layer
    dah_dzh=sigmoid_der(zh)
    #hidden layer                  
    dzh_dwh=features_set
    #input layer                  
    dcost_wh=np.dot(dzh_dwh.T, dah_dzh*dcost_dah)
    
    #Weights update
    wh-=lr*dcost_wh
    wo-=lr*dcost_wo                  
    
```
